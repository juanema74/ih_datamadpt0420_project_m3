```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import *
```

```python
diamonds = pd.read_csv('../data/diamonds_train.csv')
diamonds_predict = pd.read_csv('../data/diamonds_predict.csv')
diamonds.head()
```

```python
diamonds['vol']=diamonds['carat'].apply(lambda x: x*3,52)
```

```python
diamonds = diamonds[['carat','cut', 'color', 'depth', 'clarity', 'table', 'price', 'vol']]
diamonds.head()
```

```python
diamonds.corr()
```

```python
y = diamonds.price
```

```python
x = diamonds.drop(['price'], axis=1)
```

```python
x = pd.get_dummies(x)
```

```python
display(x.corr())
```

```python
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.pipeline import make_pipeline
pipeline = [
    StandardScaler(),
    Normalizer(),
]

tr = make_pipeline(*pipeline)

Xtest = tr.fit_transform(x)
Xtest = pd.DataFrame(Xtest,columns=x.columns)
Xtest.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(Xtest, y, test_size=0.2)
```

```python
model = GradientBoostingRegressor(n_estimators=800, max_depth=20, min_samples_leaf=3, random_state=111)
model.fit(x_train, y_train)
```

```python
print('ERROR SUMMARY')
print('MSE', mean_squared_error(y_test, model.predict(x_test)))
print('RMSE', np.sqrt(mean_squared_error(y_test, model.predict(x_test))))
print('MSLE', mean_squared_log_error(y_test, model.predict(x_test)))
print('MAE', mean_absolute_error(y_test, model.predict(x_test)))
print('R2', r2_score(y_test, model.predict(x_test)))
```
